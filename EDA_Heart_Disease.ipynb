```python
def to_cross_tab(heart1, index_name, col_name):
    df = pd.crosstab(heart1[index_name], heart1[col_name])
    df['rate'] = df.iloc[:,1] / (df.iloc[:,0] + df.iloc[:,1])
    return df
sex_target = to_cross_tab(df, 'target', 'sex')
sex_target
```
